```python
import pandas as pd
import numpy as np
import re
```

```python
converters = {'БИН': lambda x: str(x)} #in order to keep leading zeros 
df = pd.read_csv('file_converted.csv', converters = converters)
df['БИН'] = df['БИН'].astype(str)
df.head()
```

```python
df.info()
```

## Task 1

```python
#BIN should be a number consisting of 12 digits
#show incorrect bin according to the first rule
df[~df['БИН'].apply(lambda x: bool(re.search(r'\d{12}',x)))]
len(df[~df['БИН'].apply(lambda x: bool(re.search(r'\d{12}',x)))])
```

```python
#Without incorrect
df[~df['БИН'].apply(lambda x: bool(re.search(r'\d{12}',x))) == False]
```

```python
#check 2 last digits of the 1st part (registration month) < 12
df2 = df[~df['БИН'].apply(lambda x: bool(re.search(r'\d{12}',x))) == False]
df2[df2['БИН'].apply(lambda x: int(x[2:4]) > 12)]
```

```python
#Number of incorrect bins according to the month
sum(df2['БИН'].apply(lambda x: int(x[2:4]) > 12))
```

```python
#No sense to check a year, it could be the birth year of an owner etc
sum1 = sum(df['БИН'].apply(lambda x: bool(re.search(r'\d{12}',x))) == False)
sum2 = sum(df2['БИН'].apply(lambda x: int(x[2:4]) > 12))
incorrect_number = sum1 + sum2
percent_incorrect = incorrect_number/len(df)
print(f'Процентное соотношение некорректных БИН = {str(np.around(percent_incorrect, 7)*100) + "%"} ')
```

```python
#Check the 5th digit that could be only 4,5 or 6
df2_3 = df2[~df2['БИН'].apply(lambda x: int(x[2:4]) > 12)]
sum(df2_3['БИН'].apply(lambda x: x[4] != '4' and x[4] != '6' and x[4]!= '5'))
df2_3[df2_3['БИН'].apply(lambda x: x[4] != '4' and x[4] != '6' and x[4]!= '5')]
```

```python
df_incorrect1 = df[~df['БИН'].apply(lambda x: bool(re.search(r'\d{12}',x)))].set_index("БИН")
df_incorrect2 = df2[df2['БИН'].apply(lambda x: int(x[2:4]) > 12)].set_index("БИН")
df_incorrect3 = df2_3[df2_3['БИН'].apply(lambda x: x[4] != '4' and x[4] != '6' and x[4]!= '5')].set_index("БИН")
final_incorrect = pd.concat([df_incorrect1, df_incorrect2, df_incorrect3] )
final_incorrect.to_excel("Некорректные БИН.xlsx")
```

# Task 3

```python
def task3(s):
    ss = s.split(" ")
    ss = ss[-1:] + ss[:-1]
    ss = " ".join(map(str,ss))
    return ss
```

```python
df_task2['Контрагент'] = df_task2['Контрагент'].apply(lambda x: task3(x) )
df_task2
```

# Task 4

```python
def task4(b):
    l = ['ЮЛ', 'ФЛ']
    if b[4] == '4' or b[4] == '5':
        return l[0]
    return l[1]
```

```python
df_task2['ФЛ\ЮЛ'] = df_task2['БИН'].apply(lambda x: task4(x) )
df_task2
```

```python
df_task2.set_index("БИН", inplace = True)
df_task2
```
